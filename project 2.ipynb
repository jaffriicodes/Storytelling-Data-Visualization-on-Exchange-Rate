### Storytelling Data Visualization on Exchange Rates

there are two types of data visualization:
 * Exploratory data visualization: we create graphs for ourselves to better understand and explore data.
 * Explanatory data visualization: we create graphs for others to inform, make a point, or tell a story.

Throughout the Project, we focused on explanatory data visualization and learned the following:

* How to use information design principles (familiarity and maximizing the data-ink ratio) to create better graphs for an audience.
* About the elements of a story and how to create storytelling data visualizations using Matplotlib.
* How to guide the audience's attention with pre-attentive attributes.
* How to use Matplotlib built-in styles â€” with a case study on the FiveThirtyEight style.

## Rolling Mean

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_Dollar"])
plt.show
```

If we look at the line's shape, we see many small wiggles â€” rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

```python
values = pd.DataFrame()
values["daily_values"] = pd.Series(range(1,20,2))
values
```

```python
values["rolling_mean_2"] = values["daily_values"].rolling(2).mean()
values
```

```python
values["rolling_mean_3"] = values["daily_values"].rolling(3).mean()
values["rolling_mean_5"] = values["daily_values"].rolling(5).mean()
values
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_Dollar"])
plt.title("original_values", weight="bold")

for i, rolling_mean in zip([2,3,4,5,6], [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_Dollar"].rolling(rolling_mean).mean())
    plt.title("rolling_window" + str(rolling_mean), weight="bold")
    plt.yticks([1, 1.2, 1.4, 1.6])
    
plt.tight_layout() # outo adjust the subplot
plt.show()
```

## Storytelling Data Visualization

Financial Crisis 2007-2008

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >=2006) & (euro_to_dollar["Time"].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >=2007) & (euro_to_dollar["Time"].dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use("fivethirtyeight")

### adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis["Time"], financial_crisis["rolling_mean"], linewidth = 1, color = "#A6D785")

### highlight the 2007-2008 period
ax.plot(financial_crisis_7_8["Time"], financial_crisis_7_8["rolling_mean"], linewidth = 1, color = "#e23d28")

ax.set_xticklabels([])

x = 0.02
for year in ["2006", "2007", "2008", "2009", "2010"]:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform = plt.gca().transAxes)
    x += 0.22888
    
ax.set_yticklabels([])

y = 0.07
for rate in ["1.2", "1.3", "1.4", "1.5"]:
    ax.text(-.04, y, rate, alpha=0.5, fontsize=11, transform = plt.gca().transAxes)
    y += 0.2333
    
### adding a title and subtitle 
ax.text(-.05, 1.2, "Euro-USD rate peaked at 1.95 during 2007-2008 Financial Crisis", weight = "bold", transform = plt.gca().transAxes)
ax.text(-.05, 1.1, "Euro-USD exchange rate between 2006 to 2008", size=12,  transform = plt.gca().transAxes)

### Adding the signature
ax.text(-.05, -0.25, "@Techma_Zone" + " "*80 + "Source: European Cental Bank", color="#f0f0f0", backgroundcolor="#4d4d4d", size=12, transform = plt.gca().transAxes)

ax.axvspan(xmin=pd.to_datetime("2008-04-1"), xmax=pd.to_datetime("2008-09-1"), ymin=0.09, alpha=0.3, color="grey")

plt.show()
```

## The three US Presidencies Example

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2001) & 
                                        (euro_to_dollar["Time"].dt.year < 2021)]
bush = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2001) & 
                             (euro_to_dollar["Time"].dt.year < 2009)]
obama = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2009) & 
                             (euro_to_dollar["Time"].dt.year < 2017)]

trump = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2017) & 
                             (euro_to_dollar["Time"].dt.year < 2021)]
```

Below, you'll notice we used matplotlib functional approach to build the graph. We use this approach because it offer the more flexibility in arranging the subplots:

1. We first build three of the graph on a 2-by-3 grid(this grid should have six subplot, but we only build three. the bottom row will remain empty).
2. We then build only the bottom graph of a 2-by-1 grid(this grid should have two subplot, the top row will remain empty).
3. The two grid are merged, and we end upwith three graphs on the top row and one graph on the botton row.

```python
### adding the fivethirtyeight style
style.use("fivethirtyeight")

### adding the subplot
plt.figure(figsize=(12,6))

### pattern 1
ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

### pattern 2
ax4 = plt.subplot(3,1,2)
```

```python
### adding the fivethirtyeight style
style.use("fivethirtyeight")

### adding the subplot
plt.figure(figsize=(14,8))

### pattern 1
ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

### pattern 2
ax4 = plt.subplot(3,1,2)

axes = [ax1, ax2, ax3, ax4]

### change all the subplot
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(["1.0", "1.2", "1.4", "1.6 $"], alpha = 0.4)
    
### ax1: Bush
ax1.plot(bush["Time"], bush["rolling_mean"], color = "#BF5FFF")
ax1.set_xticklabels([" ", "2001", " ", "2003", " ", "2005", " ", "2007", " ", "2009"], alpha = 0.3, size=12)
ax1.text(0.11, 2.45, "BUSH", fontsize=12, weight="bold", color = "#BF5FFF", transform = plt.gca().transAxes)
ax1.text(0.08, 2.34, "(2001-2009)", weight="bold",alpha=0.3, color = "#BF5FFF", transform = plt.gca().transAxes)

### ax1: Obama
ax2.plot(obama["Time"], obama["rolling_mean"], color = "#ffa500")
ax2.set_xticklabels([" ", "2009", " ", "2011", " ", "2013", " ", "2015", " ", "2017"], alpha = 0.3, size=12)
ax2.text(0.45, 2.45, "OBAMA", fontsize=12, weight="bold", color = "#ffa500", transform = plt.gca().transAxes)
ax2.text(0.43, 2.34, "(2009-2017)", weight="bold",alpha=0.3, color = "#ffa500", transform = plt.gca().transAxes)

### ax1: trump
ax3.plot(trump["Time"], trump["rolling_mean"], color = "#00B2EE")
ax3.set_xticklabels([" ", "2017", " ", "2019", " ", "2021", " ", "2023", " ", "2025"], alpha = 0.3, size=12)
ax3.text(0.8, 2.45, "TRUMP", fontsize=12, weight="bold", color = "#00B2EE", transform = plt.gca().transAxes)
ax3.text(0.78, 2.34, "(2017-2021)", weight="bold",alpha=0.3, color = "#00B2EE", transform = plt.gca().transAxes)

### Ax4: Bush-Obama-Trump
ax4.plot(bush["Time"], bush["rolling_mean"], color = "#BF5FFF")
ax4.plot(obama["Time"], obama["rolling_mean"], color = "#ffa500")
ax4.plot(trump["Time"], trump["rolling_mean"], color = "#00B2EE")

ax4.set_xticks([])

# adding a title and subtitle
ax1.text(-0.05, 2.8, "EURO-USD rate averaged 1.22 under the last three US president", fontsize=20, weight="bold", transform = plt.gca().transAxes)
ax1.text(-0.05, 2.65, "EURO-USD exchange rate under George W.Bush (2001-2009), Barack Obama (2009-2017), Donald Trump (2017-2021)", fontsize=14, transform = plt.gca().transAxes)

# adding signature
ax4.text(-0.05, -0.15, "@TechmaZone" + " "*133 + "Source: European Central Bank", color="#f0f0f0", backgroundcolor="#4d4d4d", size=14, transform = plt.gca().transAxes)

plt.tight_layout()
plt.show()
```
